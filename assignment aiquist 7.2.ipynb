#Preprocessing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dataset= pd.read_csv('/content/drive/MyDrive/Artificial Intelligence /CoLab /Dataset/weight-height.csv')
```

```python
dataset.head(-10)
```

```python
dataset.replace(to_replace='Male',value=1,inplace=True)
dataset.replace(to_replace='Female',value=0,inplace=True)
```

```python
#dataset.head(-5)
```

```python
x=dataset.drop('Weight',axis=1)
y=dataset['Weight']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
LinReg=LinearRegression()
```

```python
LinReg.fit(xtrain,ytrain)
```

```python
ypred=LinReg.predict(xtest)
```

###Linear Reg Test scores

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.metrics import accuracy_score
```

```python
mse=mean_squared_error(ytest,ypred)
mae=mean_absolute_error(ytest,ypred)
r2e=r2_score(ytest,ypred)
rmse=np.sqrt(mse)
print("MSE: ",mse,"\nMAE: ",mae,"\nr2: ",r2e,"\nRMSE: ",rmse)
```

```python
acc_lin_reg=accuracy_score(ytest,ypred)
acc_lin_reg
```

# **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
MMSX=MinMaxScaler()
```

```python
dataset['rent-normalized']=MMSX.fit_transform(dataset[['rent']])
```

```python
dataset.head()
```

# **Standardiatio**

```python
from sklearn.preprocessing import StandardScaler
STDScaler=StandardScaler()
```

```python
dataset['rent-standardized']=STDScaler.fit_transform(dataset[['rent']])
dataset.head()
```

```python
std_ddof0=dataset.rent.std(ddof=0)
std_ddof0
```

```python
std_ddof1=dataset.rent.std(ddof=1)
std_ddof1
```

```python
std_ddof2=dataset.rent.std(ddof=2)
std_ddof2
```

```python
mean_rent=dataset.rent.mean()
mean_rent
```

```python
dataset['sdt_ddof0']=(dataset[['rent']]-mean_rent)/std_ddof0
dataset['sdt_ddof1']=(dataset[['rent']]-mean_rent)/std_ddof1
dataset['sdt_ddof2']=(dataset[['rent']]-mean_rent)/std_ddof2
dataset.head()
```

# **Log Transformation**

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
fcT=FunctionTransformer(np.log1p) #np->log1p is the natural logarithm
dataset['rent_log']=fcT.fit_transform(dataset[['rent']])
```

```python
dataset.head()
```

# **Rebust Scaler**

```python
from sklearn.preprocessing import MaxAbsScaler
mxsc=MaxAbsScaler()
```

```python
dataset['max-abs-rent']=mxsc.fit_transform(dataset[['rent']])
```

```python
dataset.head()
```

# **Power Transformer**

```python
from scipy.stats import boxcox
dataset['boxcox-rent'],_=boxcox(dataset['rent']+1)
dataset.head()
```
